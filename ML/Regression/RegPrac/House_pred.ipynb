```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
data = {

'Area':[1000,1200,1500,1800,2000,2200,2500,2700,3000,3500],

'Bedrooms':[2,2,3,3,4,4,4,5,5,6],

'Age':[12,10,8,7,5,4,3,2,2,1],

'Distance':[15,12,10,8,7,5,4,3,2,1],

'Price':[40,48,60,72,85,95,110,125,145,180]

}

df=pd.DataFrame(data)

print(df.head())
```

```python
# Step 3 dataset information

print(df.info())

print(df.describe())
```

```python
#Step 4 Correlation
print(df.corr())
```

```python
# Graph 1 Scatter Plot (Area vs Price)
plt.figure(figsize=(8,5))

plt.scatter(df['Area'],
            df['Price'],
            color='blue',
            s=120)

plt.xlabel("Area (sq.ft)")
plt.ylabel("House Price")
plt.title("Area vs House Price")

plt.grid(True)

plt.show()
```

```python
#Graph 2 Histogram
plt.figure(figsize=(8,5))

plt.hist(df['Price'],
         bins=6)

plt.xlabel("House Price")
plt.ylabel("Frequency")
plt.title("Distribution of House Price")

plt.show()
```

```python
#Graph 3 Area Distribution
plt.figure(figsize=(8,5))

plt.hist(df['Area'],
         bins=5)

plt.xlabel("Area")

plt.ylabel("Frequency")

plt.title("Distribution of Area")

plt.show()
```

```python
#Step 5 Features
X=df[['Area',
      'Bedrooms',
      'Age',
      'Distance']]

y=df['Price']
```

```python
#Step 6 Train Test Split
X_train,X_test,y_train,y_test=train_test_split(
X,
y,
test_size=0.20,
random_state=42
)
```

```python
#Step 7 Train Model
model=LinearRegression()

model.fit(X_train,y_train)
```

```python
#Prediction
y_pred=model.predict(X_test)

print(y_pred)
```

```python
#Graph 4 Actual vs Predicted 
plt.figure(figsize=(8,5))

plt.scatter(y_test,
            y_pred,
            s=150)

plt.plot([y.min(),y.max()],
         [y.min(),y.max()],
         linewidth=2)

plt.xlabel("Actual Price")

plt.ylabel("Predicted Price")

plt.title("Actual vs Predicted")

plt.grid(True)

plt.show()
```

```python
#Graph 5 Residual Plot

residuals=y_test-y_pred

plt.figure(figsize=(8,5))

plt.scatter(y_pred,
            residuals,
            s=150)

plt.axhline(y=0,
            linewidth=2)

plt.xlabel("Predicted Price")

plt.ylabel("Residual")

plt.title("Residual Plot")

plt.grid(True)

plt.show()
```

```python
#Graph 6: Feature Importance (Using Coefficients)

coef=pd.Series(model.coef_,
               index=X.columns)

coef.plot(kind='bar',
          figsize=(8,5))

plt.title("Feature Importance")

plt.ylabel("Coefficient")

plt.show()
```

```python
#Step 9: Evaluation

print("R2 Score :",r2_score(y_test,y_pred))

print("MAE :",mean_absolute_error(y_test,y_pred))

print("MSE :",mean_squared_error(y_test,y_pred))

print("RMSE :",np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
#Step 10: Predict New House
new_house=[[2800,
            4,
            2,
            3]]

prediction=model.predict(new_house)

print("Predicted Price :",prediction[0],"Lakhs")
```
