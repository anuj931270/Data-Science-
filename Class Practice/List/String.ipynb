```python
# waf to remove last word from string
# waf to remove words if vowel present from string
# waf to count total whitespace in string
# waf to extract only numbers for string
# waf to remove print all special symbols
# waf ro reverse all words of string
# waf to remove all whitespace for string
```

```python
# waf to remove last word from string
str1 = "This is python code"
str1 = str1.rsplit(" ",1)[0]
print(str1)
```

```python
# waf to remove words if vowel present from string
str1 = "This is python code"
vowel="aeiouAEIOU"
for i in str1:
    if i in vowel:
        print(i)
```

```python
# waf to count total whitespace in string
str1 = "This is python code"
count=0
for i in str1:
    if i.isspace():
        count+=1
print("total whitespace",count)
```

```python
# waf to remove all whitespace for string
str1 = "This is python code"
result = str1.replace(" ", "")
print(result)
```

```python
# waf ro reverse all words of string
str1 = "This is python code"
reverse=str1[::-1]
print(reverse)
```
