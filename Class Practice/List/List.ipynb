# Python Collections (Arrays)
There are four collection data types in the Python programming language:

    List is a collection which is ordered and changeable. Allows duplicate members.
    Tuple is a collection which is ordered and unchangeable. Allows duplicate members.
    Set is a collection which is unordered, unchangeable*, and unindexed. No duplicate members.
    Dictionary is a collection which is ordered** and changeable. No duplicate members

Negative indexing means start from the end
-1 refers to the last item, -2 refers to the second last item etc.

```python
thislist = ["apple", "banana", "cherry"]
print(thislist[-1])
```

# Range of Indexes
You can specify a range of indexes by specifying where to start and where to end the range.
When specifying a range, the return value will be a new list with the specified items.

```python
thislist = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
print(thislist[2:5])
```

```python
thislist = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
print(thislist[2:])
```

# Range of Negative Indexes
Specify negative indexes if you want to start the search from the end of the list:

```python
# This example returns the items from "orange" (-4) to, but NOT including "mango" (-1)
thislist = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
print(thislist[-4:-1])
```

# Extend List
To append elements from another list to the current list, use the extend() method.

```python
mylist = ["apple", "banana", "cherry"]
tropical = ["mango", "pineapple", "papaya",99]
mylist.extend(tropical)
print(mylist)
```

```python
mylist=["car", True, 88, "anuj"]
line=("bus", False, 748, "apple") #Add Any Iterable
mylist.extend(line)
print(mylist)
```

```python
mylist = ['apple', 'banana', 'cherry']
mylist.insert(0, 'orange')
print(mylist[1])
```

# Add Any Iterable

# Remove Specified Item
The remove() method removes the specified item.

```python
# Remove "banana":
thislist = ["apple", "banana", "cherry"]
thislist.remove("banana")
print(thislist)

# pop
thislist = ["apple", "banana", "cherry"]
thislist.pop(1)
print(thislist)

# del
thislist = ["apple", "banana", "cherry"]
del thislist[0]
print(thislist)
```

# Clear the List
The clear() method empties the list.
The list still remains, but it has no content.

```python
# Clear the list content:
thislist = ["apple", "banana", "cherry"]
thislist.clear()
print(thislist)

```

# Using a While Loop
You can loop through the list items by using a while loop.
Use the len() function to determine the length of the list, then start at 0 and loop your way through the list items by referring to their indexes.
Remember to increase the index by 1 after each iteration.

```python
# Print all items, using a while loop to go through all the index numbers
thislist = ["apple", "banana", "cherry"]
i = 0
while i < len(thislist):
  print(thislist[i])
  i = i + 1 
```

# Common List Methods & Slicing
Once you have a list, you'll want to manipulate it. Python provides built-in methods to add, remove, and slice data.

```python
nums = [10, 20, 30, 40, 50, 60]

print(nums[1:4])   # Output: [20, 30, 40] (Index 4 is excluded)
print(nums[:3])    # Output: [10, 20, 30] (From the beginning)
print(nums[::2])   # Output: [10, 30, 50] (Every second item)
print(nums[::-1])  # Output: [60, 50, 40, 30, 20, 10] (Reverse the list)
```

```python
tools = ["hammer", "screwdriver"]

# Adding items
tools.append("wrench")       # Adds to the end: ['hammer', 'screwdriver', 'wrench']
tools.insert(1, "pliers")    # Inserts at index 1: ['hammer', 'pliers', 'screwdriver', 'wrench']

# Removing items
tools.remove("pliers")       # Removes by value
popped_item = tools.pop()    # Removes and returns the last item ('wrench')

# Ordering
numbers = [4, 1, 8, 3]
numbers.sort()               # Sorts in-place: [1, 3, 4, 8]
print(tools,numbers, popped_item)
```

# List Comprehensions & Matrices
Advanced list usage often involves processing data efficiently or nesting lists inside lists.
# List Comprehensions
This is a sleek, concise way to create lists based on existing lists.

```python
# Traditional way to get squares of even numbers
squares = []
for x in range(10):
    if x % 2 == 0:
        squares.append(x**2)

# Advanced way: List Comprehension
squares_comp = [x**2 for x in range(10) if x % 2 == 0]
print(squares_comp)  # Output: [0, 4, 16, 36, 64]
```
