# Exception Handling
Exception Handling is a mechanism used to handle runtime errors in a program so that the program does not crash unexpectedly.
Exception Handling is used to handle runtime errors in a program so that the program does not crash and can continue executing normally.

```python
# Handling Specific Exceptions

try:
    num = 10 / 0

except ZeroDivisionError:
    print("Cannot divide by zero")
```

```python
# Multiple Exceptions

try:
    num = int(input("Enter a number: "))
    result = 10 / num

except ValueError:
    print("Invalid input")

except ZeroDivisionError:
    print("Cannot divide by zero")
else: ## The else block runs only when no exception occurs.
    print("Valid Input")
```

```python
# finally Block
# The finally block always executes whether an exception occurs or not

try:
    print(10 / 0)

except ZeroDivisionError:
    print("Error")

finally:
    print("This block always runs")
```

```python
mylist = [10, 20, 30]
try:
    input(print(mylist[2]))
except IndexError:
    print("Index out of range")
```
