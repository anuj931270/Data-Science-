```python
years = int(input("Enter Years of experience: "))

if years > 15:
    print("Senior Employee (Experience exceeds 15 years. Bonus added.)")
    base_salary = 80000
    bonus = 5000
    final_salary = base_salary + bonus

elif years >= 10:
    print("Senior Employee")
    final_salary = 80000

elif years >= 5:
    print("Mid-level employee")
    final_salary = 50000

else:
    print("Junior employee")
    final_salary = 30000
    print(f"Salary: {final_salary}")
```

```python
days=int(input("Enter the number of days the book has been borrowed:"))

if days <= 5:
    charge = days * 2
    print(f"The total library charge for {days} days is: Rs. {charge}")
elif days <= 10:
    charge = days * 3
    print(f"The total library charge for {days} days is: Rs. {charge}")
elif days <= 15:
    charge = days * 4
    print(f"The total library charge for {days} days is: Rs. {charge}")
else:
    charge = days * 5
    print(f"The total library charge for {days} days is: Rs. {charge}")
```
