```python
# what is HOF; a function that takes another function as argument known as HOF.
# filter(func, iterables)
```

```python
def check_pass_fail(marks):
    pass_stu=[]
    for i in marks:
        if i>55:
            pass_stu.append(i)
    return pass_stu
marks=[20,50,98,55,60,58]
res=filter(check_pass_fail,marks)
print(res)
```

```python
def check_pass_fail(marks):
    return marks >= 55
marks=[20,50,98,55,60,58]
print(list(filter(check_pass_fail,marks)))
```
