```python
pre_marks = int(input("Enter Your pre marks: "))

if pre_marks >= 300:
    print("You are pass in pre exam")
    mains_marks = int(input("Enter your mains exam: "))
    
    if mains_marks >= 500:
        print("You are pass in mains exam")
        inter_marks = int(input("Enter your interview marks: "))
        
        if inter_marks >= 700:
            print("You are selected!")
        else:
            print("You are failed in interview! Better luck next time.")
    else:
        print("You are fail in main exam")
else:
    print("You are fail in pre exam")
```
