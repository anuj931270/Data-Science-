```python
def count_vowels(text):
    count = 0

    for a in text.lower():
        if a in "aeiouAEIOU":
            count += 1
    return count
print(count_vowels("This is Python"))
```

```python
def find_vowel(text):
    all_vowels=""
    for b in text:
        if b in "aeiouAEIOU":
            all_vowels +=b
    return all_vowels
res=find_vowel("How Are You")
print("Extracted Vowels :" ,res)
```
