```python
str1 = "python programming"
vowels = "aeiouAEIOU"
size = len(str1)
i = size - 1
vowels_count = 0
count_constant= 0
while i >= 0:
    print(str1[i], end="")
    if str1[i] in vowels:
        vowels_count += 1
    else:
        count_constant +=1
    i -= 1

print(f"Total vowels: {vowels_count}")
print(f"Total constants: {count_constant}")
```
