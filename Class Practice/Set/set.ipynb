# Set
Sets are used to store multiple items in a single variable.

A set is a collection which is unordered, unchangeable*, and unindexed.

Set is a data structure used to store unique values in curly{}

Sets can be hetorgenous and homogenous

Sets are mutable in  nature

Sets always reconaize values



```python
# creation of sets

days={"monday","tuesday","wednesday","saturday","friday"}
print(days)
```

```python
#using set()
numbers = set([1, 2, 3, 4, 5])
print(numbers)
```

```python
# Duplicate Values
num = {1, 2, 2, 3, 4, 4, 5}
print(num)
```

```python
fruits = {"apple", "banana", "mango"}

print("apple" in fruits)
print("grapes" in fruits)
```

```python
fruits = {"apple", "banana", "mango"}

for item in fruits:
    print(item)
```

```python
fruits = {"apple"}

fruits.update(["banana", "mango", "orange"])

print(fruits)
```

```python
# Set Operations union()
a = {1,2,3}
b = {3,4,5}

print(a.union(b))
```

```python
# intersection() Common values.
a = {1,2,3}
b = {2,3,4}

print(a.intersection(b))
```

```python
# symmetric_difference() Elements in one set but not both.
a = {1,2,3}
b = {3,4,5}

print(a.symmetric_difference(b))
```

```python
# intersection_update() Updates original set.
a = {1,2,3}
b = {2,3,4}

a.intersection_update(b)
a.difference_update(b)
a.symmetric_difference_update(b)

print(a)
```

```python
# Find Sum of Set Elements
myset = {10, 20, 30, 40}
total=0

for i in myset:
    total+=i
print("The total sum: ", total)
```

```python
# Find Maximum Element
myset = {25, 10, 50, 15}
largest=next(iter(myset))

for i in myset:
    if i >largest:
        largest=i
print("The maximum number: ", largest)
```

```python
myset = {25, 10, 50, 15}
smallest=next(iter(myset))

for i in myset:
    if i<smallest:
        smallest=i
print("The smallest number: ",smallest)
```

```python
# Print only odd numbers from a set
myset = {1, 2, 3, 4, 5, 6, 7}
odd_count=0
for i in myset:
    if i%2!=0:
        odd_count+=1
print("The odd number: ",odd_count)
```

```python
# Print only even numbers from a set
myset = {1, 2, 3, 4, 5, 6, 7}
even_count=0
for i in myset:
    if i%2==0:
        even_count+=1
print("The odd number: ",even_count)
```
