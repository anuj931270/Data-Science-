```python
# Generate 10 coupons where the first four characters are A, B, C, and D, and the last four are any random numbers, making the total unique.

def generate_cupon():
    a_to_z="abcdefghijklmnopqrstuvwxyz"
    num="1234567890"
    char=[random.choice(a_to_z).upper() for i in range(1,5)]
    num=[random.choice(num)for i in range(1,5)]
    print("".join(char+num))

for i in range(1,10):
    generate_cupon()
```
