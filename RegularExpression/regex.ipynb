```python
# Layer 1 - Literal maching
import re
text= "This product is bad for health and negative impact"
pattern = "bad"
re.search(pattern,text)
```

```python
#Find a Word in a Sentence

text = "I love Python programming."

result = re.search("Python", text)

if result:
    print("Found")
    print(result.group())
else:
    print("Not Found")
```

```python
#Find a Number
text = "Order ID is 12345"
result = re.search(r"\d+", text)
result.group()
```

```python
#Find an Email Address

text = "Please contact us at support@gmail.com"
result = re.search(r"[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}", text)
result.group()
```

```python
#Find a Phone Number
text = "Call me at 9876543210"
result = re.search(r"\d{10}", text)
print(result.group())
```

```python
#Search Date
text = "Meeting Date: 14-07-2026"

result = re.search(r"\d{2}-\d{2}-\d{4}", text)

print(result.group())
```

```python
files = ["app.py", "config.yaml", "notes.txt", "script.sh"]

pattern = r"\.py|sh$"

for file in files:
    if re.search(pattern, file):
        print(file)
```
