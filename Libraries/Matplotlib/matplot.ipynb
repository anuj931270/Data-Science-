Matplotlib is a Python library used to create charts and graphs. It helps us visualize data so that we can understand patterns and trends easily.

we can create line cahrt, bar chart, scatter plot, pie chart, histogram.
created by jhon hunter in 2003.

MATLAB:- MATLAB is a high-level programming language and a nummerical computing environment used by engineers and scientists for data analysis, visualization, and algorithmic development.

PYPLOT:- pyplot provides a user-friendly interface for plotting.Plot is a versatile function, and will take an arbitrary number of arguments.

For every x, y pair of arguments, there is an optional third argument which is the format string that indicates the color and line type of the plot.

WHY:- We use matplotlib.pyplot because it provides simple functions to create, customize, and display graphs and charts.


```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

#First line graph
x = [1,2,3,4,5]
y = [10,20,15,25,30]

plt.plot(x, y)
plt.show()
```

```python
#A dictionary named data is created.

#np.arange() Creates numbers from 0 to 49
data = {'a': np.arange(50),
        
        #np.random.randint(0,50,50) Generates 50 random integers (used for colors)
        'c': np.random.randint(0, 50, 50),
        #Generates 50 random values from a normal distribution
        'd': np.random.randn(50)}

#Creates Y-values with random variation
data['b'] = data['a'] + 10 * np.random.randn(50)
#Converts negative values to positive
data['d'] = np.abs(data['d']) * 100#Increases marker sizes

#Draws the scatter plot.
#Uses column c for point colors
#Uses column d for point sizes

plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
#subpllots in matplotlib

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

#the figure has 1 row, 2 columns, and this plot is the first plot.
plt.subplot(1, 2, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

##the figure has 1 row, 2 columns, and this plot is the second plot.
plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
```

```python
#plot1

x=np.array([0,1,2,3])
y=np.array([2,6,1,9])

# So, 2, 1, 1 means create a grid with 2 rows and 1 column, and put the next plot in the 1st position (the top slot).
plt.subplot(2,1,1)
plt.plot(x,y)

#plot2
x=np.array([0,1,2,4])
y=np.array([10,15,20,25])


#The arguments mean: (rows, columns, plot_index)
# plt.subplot(2, 1, 2): This targets the same $2 \times 1$ grid but switches focus to the 2nd position (the bottom slot).
plt.subplot(2,1,2)
plt.plot(x,y)
plt.show()
```

```python
#Bar Chart

industry=["IT","Finance","Healthcare","Retail"]
employees=[500,300,250,450]

plt.bar(industry, employees)
plt.title("Employees by Industry")
plt.xlabel("Industry")
plt.ylabel("Employees")
plt.show()
```

```python
#pie chart

sizes=[40,30,20,10]
labels=["IT","Finance","Healthcare","Retail"]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Industry Share")
plt.show()
```
