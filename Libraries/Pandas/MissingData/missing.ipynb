```python
#finding missing data
import numpy as np 
import pandas as pd

data = {
    'A': [1, 2, np.nan, 4, 5], #np.nan means null value
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df = pd.DataFrame(data)
# df
# df.isna() #how to find null value
# df.isna().sum() #count null values
# df.isna().any() #if there is any missing values
df.dropna()  # when we are removing the missing data it is work on row basiss.
# df.dropna(thresh=1)  
```

```python
df
```

```python
#filling the missing data

df.fillna(1)
```

```python
values = {'A':0,'B':100,"C":300,'D':400}
df.fillna(value=values)
```
