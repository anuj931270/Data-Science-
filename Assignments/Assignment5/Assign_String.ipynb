```python
# Filter vowels and consonants from the string "How are you sir".

string = "How are you sir"

vowels = ""
consonants = ""

for char in string:
    if char.isalpha():  # Ignore spaces and other characters
        if char.lower() in "aeiou":
            vowels += char
        else:
            consonants += char

print("Original String :", string)
print("Vowels          :", vowels)
print("Consonants      :", consonants)
```

```python
# Convert the string "qwertyuiopasdfghjklzxcvbnm" to
# "abcdefghijklmnopqrstuvwxyz".

string = "qwertyuiopasdfghjklzxcvbnm"

source = "qwertyuiopasdfghjklzxcvbnm"
target = "abcdefghijklmnopqrstuvwxyz"

result = ""

for ch in string:
    index = source.index(ch)
    result += target[index]

print("Original String :", string)
print("Converted String:", result)
```

```python
# Count spaces, digits, alphabets, and special characters in "Python 3.9
# is awesome!!".

string = "Python 3.9 is awesome!!"

spaces = 0
digits = 0
alphabets = 0
special = 0

for ch in string:
    if ch.isalpha():
        alphabets += 1
    elif ch.isdigit():
        digits += 1
    elif ch.isspace():
        spaces += 1
    else:
        special += 1

print("Original String :", string)
print("Alphabets       :", alphabets)
print("Digits          :", digits)
print("Spaces          :", spaces)
print("Special Chars   :", special)
```

```python
# Extract substring after a specific word (e.g., "Welcome to Python
# World" → substring after "to" → "Python World").


string = "Welcome to Python World"

word = "to"

index = string.find(word)

if index != -1:
    result = string[index + len(word) + 1:]
    print("Substring after", word, ":", result)
else:
    print("Word not found")
```

```python
# Check if two strings are rotations (cyclic substrings) of each other (e.g.,
# "abcd" and "cdab" → Rotations).

string1 = input("Enter first string: ")
string2 = input("Enter second string: ")

if len(string1) == len(string2) and string2 in (string1 + string1):
    print("Strings are Rotations")
else:
    print("Strings are Not Rotations")
```

```python
# .Find the longest common substring between two strings (e.g., "abcdxyz"
# and "xyzabcd" → Longest common substring = "abcd").

string1 = "abcdxyz"
string2 = "xyzabcd"

longest = ""

for i in range(len(string1)):
    for j in range(i + 1, len(string1) + 1):
        substring = string1[i:j]
        if substring in string2 and len(substring) > len(longest):
            longest = substring

print("String 1 :", string1)
print("String 2 :", string2)
print("Longest Common Substring :", longest)
```
