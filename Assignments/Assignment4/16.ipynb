```python
# Write a Python program to take input from the user to create a list of elements.
# The user should input each element of the list one by one. Create a list with
# these elements (maximum of 5 elements).

def take_input():
    mylist1=[]
    for i in range(5):
        mylist2=input(f"Enter element: {i+1}: ")
        mylist1.append(mylist2)
    return mylist1

res=take_input()
print(res)
```

```python
# Write a Python program to generate a list of numbers in reverse order from 10 to
# 1 using list comprehension.

reverse_num=[i for i in range(10,0,-1)]
print(reverse_num)
```

```python
# Create a list of square numbers from 1 to 10 using list comprehension.

square_num=[i**2 for i in range(1,11)]
print(square_num)
```

```python
# Filter strings from the list language = ['python', 'php', 'java',
# 'c++', 'javascript', 'ruby'] that contain a specific letter provided by
# the user, using list comprehension.

lang=['python', 'php', 'java','c++', 'javascript', 'ruby']
letter=input("Enter a letter: ")

res=[lang for lang in lang if letter.lower() in lang.lower()]
if res :
    print("Languages containing", letter, ":", res)
else:
    print("User enter wrong letter")
```

```python
# Flatten a nested list like [[1,2], [3,4], [5,6]] into a single list.

nested_list=[[1,2], [3,4], [5,6]]   #for i in sublist:- takes each number from the inner list
flt_lst=[i for sublist in nested_list for i in sublist]  #takes each inner list ([1,2], [3,4], [5,6])
print(flt_lst)


nested_list=[[1,2], [3,4], [5,6]]
flt_lst=[]
for sublist in nested_list:
    for i in sublist:
        flt_lst.append(i)
print(flt_lst)
```
