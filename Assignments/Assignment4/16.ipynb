```python
# Filter strings from the list language = ['python', 'php', 'java',
# 'c++', 'javascript', 'ruby'] that contain a specific letter provided by
# the user, using list comprehension.

lang=['python', 'php', 'java','c++', 'javascript', 'ruby']
letter=input("Enter a letter: ")

res=[lang for lang in lang if letter.lower() in lang.lower()]
if res :
    print("Languages containing", letter, ":", res)
else:
    print("User enter wrong letter")
```

```python
# Flatten a nested list like [[1,2], [3,4], [5,6]] into a single list.

nested_list=[[1,2], [3,4], [5,6]]   #for i in sublist:- takes each number from the inner list
flt_lst=[i for sublist in nested_list for i in sublist]  #takes each inner list ([1,2], [3,4], [5,6])
print(flt_lst)


nested_list=[[1,2], [3,4], [5,6]]
flt_lst=[]
for sublist in nested_list:
    for i in sublist:
        flt_lst.append(i)
print(flt_lst)
```

```python
# Check whether a given list is a palindrome.
# List:
# [1, 2, 3, 2, 1]
# Output:
# Palindrome

def palindrome_check(mylist):
    plndrm=[]
    for i in range(len(mylist)-1,-1,-1):
        plndrm.append(mylist[i])
    if mylist == plndrm:
           return "Palindrome"
    else:
        return "Not Palindrome"

mylist=[1,2,3,2,1]
res=palindrome_check(mylist)
print(res)
```

```python
# .Find the longest string in a list of strings.
# List:["cat", "elephant", "dog", "hippopotamus"]
# Output: "hippopotamus"

def longest_string(mylist):
    longest=mylist[0]
    for i in mylist:
        if len(i)>len(longest):
            longest=i
    return longest
mylist=["cat", "elephant", "dog", "hippopotamus"]
res=longest_string(mylist)
print(res)
```

```python
# Count how many strings in a list start with a vowel.
# List:
# ["apple", "banana", "orange", "umbrella", "grape", "ice"]
# Output:
# 4 strings start with a vowel
# (apple, orange, umbrella, ice)

def count_vowel(mylist):
    count=0
    for i in mylist:
        if i[0].lower() in "aeiou":
            print(i)
            count += 1
    return count
mylist=["apple", "banana", "orange", "umbrella", "grape", "ice"]
res=count_vowel(mylist)
print(res,"strings start with a vowel")

```

```python
# Replace all negative numbers in a list with zero
# List:
# [5, -3, 7, -1, 0, -8, 4]
#  Output:
# [5, 0, 7, 0, 0, 0, 4]

def ne_ele(mylist):
    for i in range(len(mylist)):
        if mylist[i]<0:
            mylist[i]=0
    return mylist
mylist=[5, -3, 7, -1, 0, -8, 4]
res=ne_ele(mylist)
print(mylist)
```

```python
# Create a new list containing only prime numbers from a given list.

def prime_number(mylist):
    prime_list=[]
    for i in mylist:
        if i>1:
            is_prime=True
            for j in range(2,i):
                if i%j==0:
                    is_prime=False
            if is_prime:
                prime_list.append(i)
    return prime_list

mylist = [2, 3, 4, 5, 6, 7, 8, 11, 12, 13]
res=prime_number(mylist)
print(res)
```
