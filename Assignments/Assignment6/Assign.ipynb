```python
# Write a program to calculate the sum of all keys in the dictionary
# d = {1: 1, 2: 2, 3: 3, 4: 4}

d = {1: 1, 2: 2, 3: 3, 4: 4}
total_sum=0

for i in d:
    total_sum+=i
print("The sum of all keys is :",total_sum)
```

```python
# Write a program to calculate the sum of both keys and values in the dictionary
# d = {1: 1, 2: 2, 3: 3, 4: 4}

# Define the dictionary
d = {1: 1, 2: 2, 3: 3, 4: 4}

# Calculate the sum of keys and the sum of values, then add them together
total_sum = sum(d.keys()) + sum(d.values())

# Print the result
print(f"The sum of both keys and values is: {total_sum}")
```

```python
# Create an empty dictionary called user_data.
# Allow the user to enter key-value pairs until they choose to stop.
# Print the final dictionary.

# Create an empty dictionary
user_data = {}

while True:
    # Take key and value from the user
    key = input("Enter key: ")
    value = input("Enter value: ")

    # Store in dictionary
    user_data[key] = value

    # Ask if the user wants to continue
    choice = input("Do you want to add another key-value pair? (yes/no): ").lower()

    if choice == "no":
        break

# Print the final dictionary
print("\nFinal Dictionary:")
print(user_data)
```

```python
# Write a program to calculate the total score of all students
# student_score = {1: 44, 2: 45, 3: 55}

student_score = {1: 44, 2: 45, 3: 55}

total_sum = 0

for score in student_score.values():
    total_sum += score

print("The total score of all students is:", total_sum)
```

```python
# Write a program to find the greatest key in the dictionary
# player = {7: "Dhoni", 12: "Kohli", 9: "Rohit", 89: "Bumrah"}

player = {
    7: "Dhoni",
    12: "Kohli",
    9: "Rohit",
    89: "Bumrah"
    }
greatest=0

for key in player:
    if key>greatest:
        greatest=key
print("greatest key is: ",greatest)
```

```python
# Write a program to extract alternate key-value pairs from the dictionary
# player = {7: "Dhoni", 12: "Kohli", 9: "Rohit", 89: "Bumrah"}

player = {
    7: "Dhoni",
    12: "Kohli",
    9: "Rohit",
    89: "Bumrah"
    }

alternate={}

for i, (key, value) in enumerate(player.items()):
     if i % 2 == 0:      # Take alternate items (0, 2, ...)
        alternate[key] = value

print(alternate)

```

```python
# Write a program to find all values that start with the letter ‘K’
# player = {7: "Dhoni", 12: "Kohli", 9: "Rohit", 89: "Bumrah"}

player = {
    7: "Dhoni",
    12: "Kohli",
    9: "Rohit",
    89: "Bumrah"
    }

for i in player.values():
    if i.startswith("K"):
        print(i)
```
