```python
# search of the word "learninng" exists in the file or not
def check_for_word():
    word="learning"
    with open("practice.txt","r") as f:
        data =f.read()
        if (data.find(word)!=-1):
            print("Found")
        else:
            print("not found")
check_for_word()
```

```python
# waf to find line of the file does the word "language" occur first. print-1 if word not found

def check_for_line():
    word="learning"
    data=True
    line_no=1
    with open("practice.txt","r") as f:
        while data:
            data= f.readline()
            if(word in data):
                print(line_no)
                return
            line_no +=1
    return -1
check_for_line()
```
