# What is File Handling?

File Handling means reading data from a file and writing data into a file using Python.
It helps us store data permanently instead of keeping it only in memory.

# Examples:-
Reading a text file
Writing student records
Saving user data
Creating log files

# Syntax
file = open("filename", "readmode")  

# Read Mode (r)
Used to read existing files.

```python
f=open("sample.txt","r")
data=f.read()      #reads entire file
print(data)
print(type(data))
f.close()
```

```python
f=open("sample.txt","r")
line1=f.readline()  #rads one line at a time
print(line1)
line2=f.readline()
print(line2)
```

```python
# Reads all lines and returns a list.
file = open("sample.txt", "r")
data = file.readlines()
print(data)
file.close()
```

```python
# Using with Statement
# advantage No need to write file.close()

with open("sample.txt", "r") as file:
    data = file.read()
print(data)
```

```python
# Checking File Properties
# Once the file is open, we can check some of its properties:

f = open("sample.txt", "r")
print("Filename:", f.name)
print("Mode:", f.mode)
print("Is Closed?", f.closed)

f.close()
print("Is Closed?", f.closed)

# f.name: Returns the name of the file that was opened (in this case, "sample.txt").
# f.mode: Tells us the mode in which the file was opened. Here, it’s 'r' which means read mode.
# f.closed: Returns a boolean value- False when file is currently open otherwise True.
```

```python
# Example: Reading a File in Read Mode (r)

file = open("sample.txt", "r")
content = file.read()
print(content)
file.close()
```

```python
# Append Mode ('a')
with open('sample.txt', 'a') as file:
    file.write('\nThis is a new line.')
```

```python
# Read and Write Mode ('r+')
with open('sample.txt', 'r+') as file:
    content = file.read()
    file.write('\nThis is a new line.')
```

```python
# Write and Read Mode ('w+')
with open('sample.txt', 'w+') as file:
    file.write('Hello, world!')
    file.seek(0)
    content = file.read()
```
