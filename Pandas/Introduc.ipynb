# Pandas:-

Pandas is a python Library use to: Read data CSV files, Clean data, Analyze data, Filter and sort data, Prepare data for machine learning.



```python
# series:- A series is a one dimesional array with labels (called an index).

import pandas as pd

numbers = pd.Series([10, 20, 30, 40, 50])

print(numbers)

```

```python
# Data Frame:- A DataFrame is a two-dimensional table with rows and columns.

import pandas as pd

data = {
    "Name": ["Anuj", "Rahul", "Priya"],
    "Age": [22, 25, 21],
    "City": ["Delhi", "Noida", "Pune"],
    "Salary": [1000,2000,3000]
}

df = pd.DataFrame(data)

print(df)
```
