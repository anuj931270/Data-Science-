```python
import numpy as np

data = np.array([10,20,20,30,40,50,60])

print("Data:", data)
print("Mean:", np.mean(data))
print("Median:", np.median(data))
print("Variance:", np.var(data))
print("Sample Variance:", np.var(data, ddof=1))
print("Standard Deviation:", np.std(data))
print("Sample Standard Deviation:", np.std(data, ddof=1))
print("Minimum:", np.min(data))
print("Maximum:", np.max(data))
print("Range:", np.max(data)-np.min(data))
print("25th Percentile:", np.percentile(data,25))
print("75th Percentile:", np.percentile(data,75))
print("Sum:", np.sum(data))

values, counts = np.unique(data, return_counts=True)
mode = values[np.argmax(counts)]
print("Mode:", mode)
```

```python
#Mean

import numpy as np

data = np.array([10, 20, 30, 40, 50])

total = 0

for num in data:
    total += num

mean = total / len(data)

print("Mean:", mean)
```

```python
#2. Maximum 
import numpy as np

data = np.array([10, 20, 30, 40, 50])

maximum = data[0]

for num in data:
    if num > maximum:
        maximum = num

print("Maximum:", maximum)
```

```python
#Minimum
import numpy as np

data = np.array([10, 20, 30, 40, 50])

minimum = data[0]

for num in data:
    if num < minimum:
        minimum = num

print("Minimum:", minimum)
```

```python
import numpy as np

data = np.array([10, 20, 30, 40, 50])

# Mean
total = 0
for num in data:
    total += num

mean = total / len(data)

# Variance
sum_sq = 0
for num in data:
    sum_sq += (num - mean) ** 2

variance = sum_sq / len(data)      # Population Variance

print("Variance:", variance)
```

```python
#Standard Deviation

import numpy as np

data = np.array([10, 20, 30, 40, 50])

total = 0
for num in data:
    total += num

mean = total / len(data)

sum_sq = 0
for num in data:
    sum_sq += (num - mean) ** 2

variance = sum_sq / len(data)

std = variance ** 0.5

print("Standard Deviation:", std)
```

```python
#Mode

data = [10, 20, 20, 30, 30, 30, 40]

frequency = {}

for num in data:
    if num in frequency:
        frequency[num] += 1
    else:
        frequency[num] = 1

mode = None
max_count = 0

for key in frequency:
    if frequency[key] > max_count:
        max_count = frequency[key]
        mode = key

print("Mode =", mode)
```
