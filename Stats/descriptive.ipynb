```python
# # Descriptive
# The value that represents all the data points in our dataset.

# >Mean:-
#      >Arithmetic Mean
#      >Geometric Mean
#      >Harmonic Mean
# >Median:-
#      >First(Low) Median
#      >Second Median
#      >Third(High) Median
# >Mode
#      >Single Mode
#      >Multi Mode
```

```python
import statistics
x=[10,20,30,40]
print(statistics.mean(x))  #a+b+c......n/n
print(statistics.geometric_mean(x)) #(a*b*c.....n)/1/n
print(statistics.harmonic_mean(x))  #n/1/a+1/b.....
```

```python
import statistics
import random
for i in range(50):
    x.append(random.randint(1,50))
print(x)
print(statistics.median(x))
print(statistics.quantiles(x))  #1st(Q1),2nd(Q2),3rd(Q3)
```

```python
import random
for i in range(200):
    x.append(random.randint(1,200))
print(x)
print(statistics.median(x))
print(statistics.quantiles(x))  #1st(Q1),2nd(Q2),3rd(Q3)
```

```python
x=[10,20,30,40,10,52,10,20,20,40,10]
print(statistics.mode(x))
print(x)
```
