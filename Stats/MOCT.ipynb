# Measure of Central Tendency

Let's write a python program

1.Generate 100000 random numbers.


```python
import random
import time
import numpy as np
from collections import Counter  #Counter is a class in Python that counts how many times each element appears in a list.
from scipy import stats #for mode

#Generate 10,00,000 random integers between 1 and 1000
data=[random.randint(1,1000) for _ in range (1000000)]
np_data=np.array(data)

def mean_from_scratch(arr):
    return sum(arr)/len(arr)

def median_from_scratch(arr):
    sorted_arr=sorted(arr)
    n =len(sorted_arr)
    mid=n//2
    if n%2==0:
        return (sorted_arr[mid-1]+sorted_arr[mid])/2
    else:
        return sorted_arr[mid]
    
def mode_from_scratch(arr):
    freq= Counter(arr)
    max_freq=max(freq.values())
    modes= [i for i, j in freq.items() if j== max_freq]
    return modes[0] if len(modes) ==1 else modes  

#Timing scratch
start=time.time()
mean_s=mean_from_scratch(data)
median_s=median_from_scratch(data)
mode_s=mode_from_scratch(data)
end=time.time()
scratch_time=end-start

#Timing scratch
start=time.time()
mean_np=np.mean(np_data)
median_np=np.median(np_data)
mode_np=stats.mode(np_data, keepdims=True).mode[0] #scipy mode as numpy
end=time.time()
numpy_time=end-start


print("From Scratch:")
print(f"Mean     : {mean_s}")
print(f"Median   : {median_s}")
print(f"Mode     : {mode_s}")
print(f"Time     : {scratch_time:.2f} sec\n")

print("Using Numpy:")
print(f"Mean    : {mean_np}")
print(f"Median  : {median_np}")
print(f"Mode    : {mode_np}")
print(f"Time    : {numpy_time:.6f} sec\n")

print("Speedup (Scratch / Numpy):", scratch_time / numpy_time)
```
