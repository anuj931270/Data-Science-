```python
import numpy as np #using Numpy Lib

x = np.array([10, 20, 30, 40])

# Arithmetic Mean
print("Arithmetic Mean:", np.mean(x))

# Geometric Mean
print("Geometric Mean:", np.prod(x) ** (1 / len(x)))

# Harmonic Mean
print("Harmonic Mean:", len(x) / np.sum(1 / x))
```

```python
import numpy as np
import random

# Create an empty list
x = []

# Add 50 random numbers between 1 and 50
for i in range(50):
    x.append(random.randint(1, 50))

# Convert list to NumPy array
x = np.array(x)

# Print the array
print(x)

# Median
print("Median:", np.median(x))

# Quartiles (same as statistics.quantiles)
print("Quartiles (25%, 50%, 75%):", np.quantile(x, [0.25, 0.50, 0.75]))
```
